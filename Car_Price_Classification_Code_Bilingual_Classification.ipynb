# Import Classification Libraries

### العربية
نقوم هنا باستيراد المكتبات والنماذج الخاصة بالتصنيف، بالإضافة إلى مقاييس التقييم التي تساعدنا على مقارنة أداء النماذج المختلفة.

### English
Here, we import the classification libraries and models, as well as the evaluation metrics that help us compare the performance of different models.


```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

# Train Test Split for Classification

### العربية
بعد إنشاء فئات السعر، نقسم البيانات إلى بيانات تدريب وبيانات اختبار. بيانات التدريب تستخدم لتعليم النماذج، وبيانات الاختبار تستخدم لقياس أداء النموذج على بيانات لم يرها من قبل.

### English
After creating the price categories, we split the data into training and testing sets. The training data is used to train the models, while the testing data is used to evaluate model performance on unseen data.


```python
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_classification,
    y_classification,
    test_size=0.2,
    random_state=42,
    stratify=y_classification
)

print("Training shape:", X_train_cls.shape)
print("Testing shape:", X_test_cls.shape)
```

# Scaling Classification Features

### العربية
نقوم هنا بتوحيد قيم الخصائص باستخدام StandardScaler. هذه الخطوة مهمة خصوصًا لنماذج مثل Logistic Regression و KNN و SVM لأنها تتأثر باختلاف مقاييس الأعمدة.

### English
Here, we standardize the feature values using StandardScaler. This step is especially important for models such as Logistic Regression, KNN, and SVM because they are affected by differences in feature scales.


```python
classification_scaler = StandardScaler()

X_train_cls_scaled = classification_scaler.fit_transform(X_train_cls)
X_test_cls_scaled = classification_scaler.transform(X_test_cls)
```

# Build Classification Models

### العربية
في هذه الخطوة ننشئ مجموعة من نماذج التصنيف المختلفة حتى نتمكن من تدريبها ومقارنة نتائجها. استخدمنا Logistic Regression و Decision Tree و Random Forest و KNN و SVM و Gradient Boosting.

### English
In this step, we create several classification models so we can train them and compare their results. We use Logistic Regression, Decision Tree, Random Forest, KNN, SVM, and Gradient Boosting.


```python
classification_models = {

    "Logistic Regression": LogisticRegression(
        max_iter=1000,
        random_state=42
    ),

    "Decision Tree Classifier": DecisionTreeClassifier(
        max_depth=10,
        random_state=42
    ),

    "Random Forest Classifier": RandomForestClassifier(
        n_estimators=300,
        max_depth=15,
        min_samples_split=5,
        random_state=42
    ),

    "KNN Classifier": KNeighborsClassifier(
        n_neighbors=7
    ),

    "SVM Classifier": SVC(
        kernel="rbf",
        C=100,
        gamma="scale",
        random_state=42
    ),

    "Gradient Boosting Classifier": GradientBoostingClassifier(
        n_estimators=200,
        learning_rate=0.05,
        random_state=42
    )
}
```

# Train and Evaluate Classification Models

### العربية
نقوم هنا بتدريب كل نموذج على بيانات التدريب ثم التنبؤ بفئات الأسعار في بيانات الاختبار. بعد ذلك نحسب Accuracy و Precision و Recall و F1 Score لمقارنة أداء النماذج.

### English
Here, we train each model on the training data and then predict the price categories for the testing data. Then we calculate Accuracy, Precision, Recall, and F1 Score to compare the models' performance.


```python
classification_results = []

for name, model in classification_models.items():

    print(f"Training Model: {name}")
    print("-" * 50)

    model.fit(X_train_cls_scaled, y_train_cls)

    y_pred_cls = model.predict(X_test_cls_scaled)

    accuracy = accuracy_score(y_test_cls, y_pred_cls)
    precision = precision_score(y_test_cls, y_pred_cls, average="weighted", zero_division=0)
    recall = recall_score(y_test_cls, y_pred_cls, average="weighted", zero_division=0)
    f1 = f1_score(y_test_cls, y_pred_cls, average="weighted", zero_division=0)

    classification_results.append({
        "Model": name,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1
    })

    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1)
    print()
```

# Compare Classification Models

### العربية
نحول نتائج النماذج إلى جدول مرتب حسب F1 Score. هذا يساعدنا على معرفة أفضل نموذج بشكل واضح، لأن F1 Score يجمع بين Precision و Recall في مقياس واحد.

### English
We convert the model results into a table sorted by F1 Score. This helps us clearly identify the best model because F1 Score combines Precision and Recall into one metric.


```python
classification_results_df = pd.DataFrame(classification_results)
classification_results_df = classification_results_df.sort_values(
    by="F1 Score",
    ascending=False
)

classification_results_df
```

# Select the Best Classification Model

### العربية
بعد مقارنة النماذج، نختار أفضل نموذج بناءً على أعلى قيمة F1 Score. ثم نعيد تدريبه أو نستخدمه مباشرة لاستخراج تقرير التصنيف ومصفوفة الالتباس.

### English
After comparing the models, we select the best model based on the highest F1 Score. Then we use it to generate the classification report and the confusion matrix.


```python
best_classification_model_name = classification_results_df.iloc[0]["Model"]
best_classification_model = classification_models[best_classification_model_name]

print("Best Classification Model:", best_classification_model_name)
```

# Classification Report and Confusion Matrix

### العربية
في هذه الخطوة نعرض تقرير التصنيف للنموذج الأفضل. يوضح التقرير أداء النموذج في كل فئة من فئات السعر، بينما توضح مصفوفة الالتباس عدد التوقعات الصحيحة والخاطئة لكل فئة.

### English
In this step, we display the classification report for the best model. The report shows the model's performance for each price category, while the confusion matrix shows the number of correct and incorrect predictions for each class.


```python
best_y_pred_cls = best_classification_model.predict(X_test_cls_scaled)

print("Classification Report")
print(classification_report(y_test_cls, best_y_pred_cls, zero_division=0))

print("Confusion Matrix")
print(confusion_matrix(y_test_cls, best_y_pred_cls))
```

# Visualize the Confusion Matrix

### العربية
نستخدم هذا الرسم لتوضيح مصفوفة الالتباس بشكل بصري. القيم الموجودة على القطر الرئيسي تمثل التوقعات الصحيحة، وكلما زادت هذه القيم كان أداء النموذج أفضل.

### English
We use this plot to visualize the confusion matrix. The values on the main diagonal represent correct predictions, and the higher these values are, the better the model performs.


```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(y_test_cls, best_y_pred_cls)

display = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=best_classification_model.classes_
)

display.plot()
plt.title("Confusion Matrix - Best Classification Model")
plt.show()
```
